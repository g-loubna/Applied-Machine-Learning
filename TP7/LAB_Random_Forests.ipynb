```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

# **Params of RFClassifier**
class sklearn.ensemble.RandomForestClassifier(n_estimators=100, *, criterion='gini', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='sqrt', max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False, class_weight=None, ccp_alpha=0.0, max_samples=None, monotonic_cst=None)

**Parameters:**
- n_estimators : int, default=100
The number of trees in the forest.
- criterion{“gini”, “entropy”, “log_loss”}: default=”gini”
The function to measure the quality of a split. Supported criteria are “gini” for the Gini impurity and “log_loss” and “entropy” both for the Shannon information gain.
- max_depth: Limits the maximum depth of the tree. If None, trees expand until all leaves are pure or contain fewer samples than min_samples_split.

- min_samples_split: The minimum number of samples required to split an internal node. It can be an integer (absolute number) or a float (fraction of the samples).

- min_samples_leaf: Minimum number of samples required to be at a leaf node. It can be an integer or a float. Helps in smoothing the model, especially in regression.

- min_weight_fraction_leaf: The minimum weighted fraction of the total sum of weights required to be at a leaf node.

- max_features: The number of features to consider when looking for the best split. It can be an integer, float, or specific values like "sqrt" or "log2".

- max_leaf_nodes: Limits the number of leaf nodes. Trees are grown in a best-first fashion.

- min_impurity_decrease: A node will be split if this split induces a decrease in impurity greater than or equal to this value.

- bootstrap: Whether bootstrap samples are used when building trees. If False, the entire dataset is used.

- oob_score: Whether to use out-of-bag samples to estimate the generalization score. Available only if bootstrap=True.

- n_jobs: Number of parallel jobs to run. -1 means using all processors.

- random_state: Controls randomness in the bootstrapping of samples and the sampling of features for each split.

- verbose: Controls the verbosity during fitting and predicting.

- warm_start: Reuse the solution of the previous fit and add more estimators to the ensemble if True.

- class_weight: Weights associated with classes to handle imbalanced datasets. Can be balanced, balanced_subsample, or custom weights.

- ccp_alpha: Complexity parameter for Minimal Cost-Complexity Pruning. Controls the trade-off between model complexity and accuracy.

- max_samples: Number of samples to draw for training each base estimator when bootstrap=True.

- monotonic_cst: Monotonicity constraint to enforce on each feature. Not supported for multiclass, multioutput classifications, or data with missing values.

```python
from sklearn.ensemble import RandomForestClassifier
RFmodel = RandomForestClassifier()
RFmodel.fit(X_train, y_train)
# Hyperparameters for a random forest
# Random Forest has various hyperparameters like n_estimators, max_depth, min_samples_split, etc.

#Compute the score for both training and testing
train_score = RFmodel.score(X_train, y_train)
test_score = RFmodel.score(X_test, y_test)
print("Training score:", train_score)
print("Testing score:", test_score)
```

```python
#Generate confusion matrix for testing data
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Predictions on test set
y_pred = RFmodel.predict(X_test)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
# Step after: Choose the best hyperparameters
# Let's use GridSearchCV to find the best hyperparameters
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [3, 5, 7],
    'n_estimators': list(range(10, 200, 10))
}

grid_search = GridSearchCV(estimator=RandomForestClassifier(), param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

print("Best hyperparameters:", grid_search.best_params_)
```

```python
# Interpret the model - Feature Importance
importance = pd.DataFrame({'Importance': RFmodel.feature_importances_ * 100}, index=load_breast_cancer().feature_names)
importance.sort_values(by='Importance', axis=0, ascending=True).plot(kind='barh', color='r')
plt.xlabel('Feature Importance')
plt.gca().legend_ = None
plt.show()
```

**Do the most Important features change if you choose a different number of trees in a random forest ?**

```python
# TO Determine if the most important features change with different number of trees
importance_dict = {}

# Iterate over different values of n_estimators
for n_estimators in range(10, 200, 10):
    # Train a Random Forest model with current n_estimators
    RFmodel = RandomForestClassifier(n_estimators=n_estimators)
    RFmodel.fit(X_train, y_train)

    # Get feature importances
    importance = pd.DataFrame({'Importance': RFmodel.feature_importances_ * 100}, index=load_breast_cancer().feature_names)
    importance_dict[n_estimators] = importance

# Plotting
fig, axes = plt.subplots(nrows=7, ncols=3, figsize=(15, 25))
for i, (n_estimators, importance_df) in enumerate(importance_dict.items()):
    row = i // 3
    col = i % 3
    ax = axes[row, col]
    importance_df.sort_values(by='Importance', axis=0, ascending=True).plot(kind='barh', color='r', ax=ax)
    ax.set_title(f'n_estimators = {n_estimators}')
    ax.set_xlabel('Feature Importance')
    ax.legend().remove()

plt.tight_layout()
plt.show()
```

While the most important features in random forest model typically stay relatively stable as the number of trees increases, the change of trees number can lead to changing and variation of feature importance.
